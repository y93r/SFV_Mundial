```python
import pandas as pd
import numpy as np

#arquivo csv é necessário fazer alguns tratamentos: 
#primeiro separador de colunas é ','; 
#descobrir o encoding; 
#trocar a ',' na coluna altura por '.'
#tirar o espaço da coluna "AlTURA"
mundial_18 = pd.read_csv(r'Mundial 2018.csv',sep= ',',encoding='latin1',skipinitialspace = True, thousands = '.', decimal = ',', dtype ={'ALTURA':np.float64})
display(mundial_18)
```

```python
#contagem de idade e altura
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
mundial_18['IDADE'].hist(bins=7, rwidth=0.50, color="darkblue")
plt.xlim(20,35) 
plt.ylim(0,5)
plt.title("Mundial 2018 - Idade", fontsize = 16)

plt.figure(figsize=(15,5))
mundial_18['ALTURA'].hist(bins=7, rwidth=0.90, color="purple")
plt.xlim(1.60,2.00) 
plt.ylim(0,7)
plt.title("Mundial 2018 - Altura", fontsize = 16)
```

```python
#criando coluna para mostrar novatas e experientes(novatas <= 25)
import sys

mundial_18['FAIXA ETÁRIA']=pd.cut(
    mundial_18['IDADE'],
    bins=[0, 25, sys.maxsize],
    labels=['Novata', 'Experiente']

)

mundial_18
```

```python
mundial_18['FAIXA ETÁRIA'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle=190)
plt.legend(mundial_18['FAIXA ETÁRIA'].value_counts(), title="Qtde em nº",  bbox_to_anchor=(1, 1))
plt.axis('equal')
plt.tight_layout()
```
