```python
#convocando as jogadoras com todas as informações necessárias
jogadoras = []

for convocadas in range (14):
    nome = input("Nome da jogadora: ").title().strip()
    numero = int(input('Nº da jogadora: '))
    idade = int(input('Idade : '))
    altura = float(input ('Altura: '))
    posicao = input ('Posição: ').title().strip()
    participacao = input ('Participação em mundiais: ').title().strip()
    convocacao = [numero, nome, idade, altura, posicao, participacao]
    jogadoras.append(convocacao)
```

```python
#criando dataframe com as convocadas 
import pandas as pd

df = pd.DataFrame(jogadoras, columns=['Nº', 'NOME', 'IDADE', 'ALTURA', 'POSIÇÃO', 'PARTICIPAÇÃO EM MUNDIAIS'])
df=df.sort_values(by =['Nº'])
display(df)

#criando dataframe com média de altura e idade
media_idade = 0
media_altura = 0
media = []

for nome, numero, idade, altura, posicao, participacao in jogadoras:
    media_idade += idade/14
    media_altura += altura/14
media = [media_idade, media_altura]


dfnew= pd.DataFrame([media], columns=['MÉDIA IDADE', 'MÉDIA ALTURA'] )
display(dfnew.round(2))                                    
```

```python
#resumo das estatísticas
display(df.describe())
```

```python
#contagem de idade e altura
import matplotlib.pyplot as plt

df=pd.DataFrame(jogadoras,columns=['Nº', 'NOME', 'IDADE', 'ALTURA', 'POSIÇÃO', 'PARTICIPAÇÃO EM MUNDIAIS'])
plt.figure(figsize=(15,5))
df['IDADE'].hist(bins=7, rwidth=0.50, color="darkblue")
plt.xlim(17,43) 
plt.ylim(0,5)
plt.title("Mundial 2022 - Idade", fontsize = 16)

plt.figure(figsize=(15,5))
df['ALTURA'].hist(bins=7, rwidth=0.90, color="purple")
plt.xlim(1.65,1.95) 
plt.ylim(0,7)
plt.title("Mundial 2022 - Altura", fontsize = 16)
```

```python
#particapações em mundiais
part_mundiais = df['PARTICIPAÇÃO EM MUNDIAIS'].value_counts()
print(part_mundiais)
```

```python
df['PARTICIPAÇÃO EM MUNDIAIS'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle=180)
plt.legend(part_mundiais, title="Qtde em nº",  bbox_to_anchor=(1, 1))
plt.axis('equal')
plt.tight_layout()
```

```python
#média de idade por posição
mundial = pd.read_excel('Mundial 2022.xlsx')
mundial.groupby('POSIÇÃO')['IDADE'].mean()
media_idade = sns.barplot(x='POSIÇÃO', y='IDADE', data=mundial, ci=None)
media_idade.set_ylim(20,34)
```

```python
#média de altura por posição
mundial.groupby('POSIÇÃO')['ALTURA'].mean()
media_altura = sns.barplot(x='POSIÇÃO', y='ALTURA', data=mundial, ci=None)
media_altura.set_ylim(1.60, 1.95)
```

```python
#criando coluna para mostrar novatas e experientes(novatas <= 25)
import sys

mundial['FAIXA ETÁRIA']=pd.cut(
    mundial['IDADE'],
    bins=[0, 25, sys.maxsize],
    labels=['Novata', 'Experiente']

)

mundial
```

```python
#Relação entre novatas e experientes e particpações em Mundiais
mundial.groupby('FAIXA ETÁRIA')['PARTICIPAÇÃO EM MUNDIAIS'].value_counts()
```

```python
import matplotlib.pyplot as plt
mundial['FAIXA ETÁRIA'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle=190)
plt.legend(mundial['FAIXA ETÁRIA'].value_counts(), title="Qtde em nº",  bbox_to_anchor=(1, 1))
plt.axis('equal')
plt.tight_layout()
```
