```python
# Analise idade por posição
analise_idade = df.groupby( by=['POSIÇÃO'] ).agg(
    Quantidade = ('IDADE', 'count'),
    Media = ('IDADE', 'mean'),
    Max = ('IDADE', 'max'),
    Min = ('IDADE', 'min'),
    Mediana = ('IDADE', 'median'),
)


analise_idade.sort_values('Media', ascending=False ).round(2)
```

```python
# Analise idade por país
analise_idade_pais = df.groupby( by=['MUNDIAL'] ).agg(
    Quantidade = ('IDADE', 'count'),
    Media = ('IDADE', 'mean'),
    Max = ('IDADE', 'max'),
    Min = ('IDADE', 'min'),
    Mediana = ('IDADE', 'median'),
)


analise_idade_pais.sort_values('Media', ascending=False ).round(2)
```

```python
# Analise altura por posição
analise_altura = df.groupby( by=['POSIÇÃO'] ).agg(
    Quantidade = ('ALTURA', 'count'),
    Media = ('ALTURA', 'mean'),
    Max = ('ALTURA', 'max'),
    Min = ('ALTURA', 'min'),
    Mediana = ('ALTURA', 'median'),
)


analise_altura.sort_values('Media', ascending=False ).round(2)
```

```python
# Analise altura por altura
analise_altura_pais = df.groupby( by=['MUNDIAL'] ).agg(
    Quantidade = ('ALTURA', 'count'),
    Media = ('ALTURA', 'mean'),
    Max = ('ALTURA', 'max'),
    Min = ('ALTURA', 'min'),
    Mediana = ('ALTURA', 'median'),
)


analise_altura_pais.sort_values('Media', ascending=False).round(2)
```
